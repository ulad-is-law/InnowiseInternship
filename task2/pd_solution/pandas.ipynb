```python
import pandas as pd
```

```python
def load_data(file_path: str) -> pd.DataFrame:
    try:
        raw_df = pd.read_excel(file_path)
    except FileNotFoundError:
        print(f"Error, file not found on the path: {file_path}")
        return pd.DataFrame()
    return raw_df

raw_df = load_data('task_2_data_ex.xlsx')
```
